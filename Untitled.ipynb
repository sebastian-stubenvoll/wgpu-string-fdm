```python
d = 0.5e-3                   # winding wire diameter (m)
D = 1.1e-3                   # core wire diameter (m)
alpha = 0.01                 # helix angle
L = 0.8
N = 128


R = D/2
h = np.tan(alpha) * R

turns = L / h
turns_per_node = int(turns / N)
turns = N * turns_per_node

L = (int(turns) * h) #adjust L fornow, so we have a whole number of turns


p = 2 * np.pi * h
kappa = R / (R**2 + h**2) #curvature
tau = h / (R**2 + h**2) #torsion
S = L * np.sqrt(R**2 + h**2) / h

turns = L / h
dt = 1e-3
ds = turns_per_node * np.sqrt((2 * np.pi * R)**2 + p**2)

oversampling_factor = 256

def winding_properties(d, E, G, rho):

    r = d / 2
    A = np.pi * r**2
    I1 = I2 = (np.pi * r**4) / 4   # bending moments of inertia
    J = (np.pi * r**4) / 2         # polar moment of inertia for torsion

    # Radii of gyration
    k1 = np.sqrt(I1 / A)
    k2 = np.sqrt(I2 / A)
    kT = np.sqrt(J / A)

    # Flexural rigidities (bending about two axes) and torsional rigidity
    beta1 = E * I1
    beta2 = E * I2
    betaT = G * J

    # Shear rigidities (sigma1, sigma2) and extensional rigidity (sigma_p)
    sigma1 = G * A
    sigma2 = G * A
    sigma_p = E * A

    mass_per_length = rho * A

    return {
        'm': mass_per_length,
        'k': [k1, k2, kT],
        'beta': [beta1, beta2, betaT],
        'sigma': [sigma1, sigma2, sigma_p]
    }


loss = 1.0
rho = 8900
E=1.1
G=1.1

coil_props = winding_properties(d, E, G, rho)
print(coil_props)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example constants (define globally or pass as args)
coil_props = {
    'm': 1.0,
    'k': [1.0, 1.0, 1.0]  # [k1, k2, kT]
}
tau = 0.1
kappa = 0.05


class Node:
    """Node with 3 translational and 3 rotational DOFs."""
    def __init__(self, s, m=1.0, k1=1.0, k2=1.0, kT=1.0):
        self.s = s
        # Translations
        self.u = self.v = self.w = 0.0
        self.u_dot = self.v_dot = self.w_dot = 0.0
        self.u_ddot = self.v_ddot = self.w_ddot = 0.0

        # Rotations
        self.th1 = self.th2 = self.thT = 0.0
        self.th1_dot = self.th2_dot = self.thT_dot = 0.0
        self.th1_ddot = self.th2_ddot = self.thT_ddot = 0.0

        # Forces and moments
        self.M1 = self.M2 = self.MT = 0.0
        self.Q1 = self.Q2 = self.P = 0.0

        # Inertia and mass
        self.m = coil_props['m']
        self.k1 = coil_props['k'][0]
        self.k2 = coil_props['k'][1]
        self.kT = coil_props['k'][2]


class BeamSimulation:
    def __init__(self, ds, n=20, L=1.0, dt=1e-3, BC_left='clamped', BC_right='free'):
        self.n = n
        self.L = L
        self.dt = dt
        self.ds = ds
        self.nodes = [Node(i * self.ds) for i in range(n)]
        self.tau = tau
        self.kappa = kappa
        self.BC_left = BC_left
        self.BC_right = BC_right

    # ------------------------------------------------------------
    #   Finite difference derivative with BCs
    # ------------------------------------------------------------
    def d_ds(self, arr):
        d = np.zeros_like(arr)
        # central difference
        d[1:-1] = (arr[2:] - arr[:-2]) / (2 * self.ds)

        # forward/backward difference near ends
        d[0] = (arr[1] - arr[0]) / self.ds
        d[-1] = (arr[-1] - arr[-2]) / self.ds
        return d

    # ------------------------------------------------------------
    #   Apply boundary conditions
    # ------------------------------------------------------------
    def apply_BCs(self):
        left = self.nodes[0]
        right = self.nodes[-1]

        # Left: Clamped (u=v=w=θ1=θ2=θT=0)
        if self.BC_left.lower() == 'clamped':
            left.u = left.v = left.w = 0.0
            left.u_dot = left.v_dot = left.w_dot = 0.0
            left.th1 = left.th2 = left.thT = 0.0
            left.th1_dot = left.th2_dot = left.thT_dot = 0.0

        # Right: Free -> zero moment and shear (already natural BC, nothing forced)
        if self.BC_right.lower() == 'free':
            pass

        # Could add: 'simply_supported', 'pinned', etc.

    # ------------------------------------------------------------
    #   Single time step
    # ------------------------------------------------------------
    def step(self):
        u = np.array([n.u for n in self.nodes])
        v = np.array([n.v for n in self.nodes])
        w = np.array([n.w for n in self.nodes])
        th1 = np.array([n.th1 for n in self.nodes])
        th2 = np.array([n.th2 for n in self.nodes])
        thT = np.array([n.thT for n in self.nodes])

        du_ds, dv_ds, dw_ds = self.d_ds(u), self.d_ds(v), self.d_ds(w)
        dth1_ds, dth2_ds, dthT_ds = self.d_ds(th1), self.d_ds(th2), self.d_ds(thT)

        # ---- (49) Moments ----
        M1 = dth1_ds + th2
        M2 = dth2_ds - self.tau * th1 + self.kappa * thT
        MT = dthT_ds - self.kappa * th2

        # ---- (50) Shear Forces ----
        Q1 = du_ds - self.tau * v + self.kappa * w - th1
        Q2 = dv_ds + self.tau * u + thT
        P  = dw_ds - self.kappa * u

        # ---- Store ----
        for i, n in enumerate(self.nodes):
            n.M1, n.M2, n.MT = M1[i], M2[i], MT[i]
            n.Q1, n.Q2, n.P  = Q1[i], Q2[i], P[i]

        # ---- (51) Rotational accelerations ----
        dM1_ds, dM2_ds, dMT_ds = self.d_ds(M1), self.d_ds(M2), self.d_ds(MT)

        for i, n in enumerate(self.nodes):
            n.th1_ddot = (dM1_ds[i] + Q1[i] + self.tau * M2[i]) / (n.m * n.k1**2)
            n.th2_ddot = (dM2_ds[i] - Q2[i] - self.tau * M1[i] + self.kappa * MT[i]) / (n.m * n.k2**2)
            n.thT_ddot = (dMT_ds[i] - self.kappa * M2[i]) / (n.m * n.kT**2)

        # ---- (52) Translational accelerations ----
        dQ1_ds, dQ2_ds, dP_ds = self.d_ds(Q1), self.d_ds(Q2), self.d_ds(P)

        for i, n in enumerate(self.nodes):
            n.u_ddot = (dQ1_ds[i] - self.tau * Q2[i] + self.kappa * P[i]) / n.m
            n.v_ddot = (dQ2_ds[i] + Q1[i]) / n.m
            n.w_ddot = (dP_ds[i] - self.kappa * Q1[i]) / n.m

        # ---- Integrate ----
        for n in self.nodes:
            n.u_dot += n.u_ddot * self.dt
            n.v_dot += n.v_ddot * self.dt
            n.w_dot += n.w_ddot * self.dt
            n.th1_dot += n.th1_ddot * self.dt
            n.th2_dot += n.th2_ddot * self.dt
            n.thT_dot += n.thT_ddot * self.dt

            n.u += n.u_dot * self.dt
            n.v += n.v_dot * self.dt
            n.w += n.w_dot * self.dt
            n.th1 += n.th1_dot * self.dt
            n.th2 += n.th2_dot * self.dt
            n.thT += n.thT_dot * self.dt

        # ---- Apply BCs after update ----
        self.apply_BCs()

    # ------------------------------------------------------------
    #   Run simulation
    # ------------------------------------------------------------
    def run(self, steps=500):
        u_hist = []
        for _ in range(steps):
            self.step()
            u_hist.append([n.u for n in self.nodes])
        return np.array(u_hist)


```

```python
# === Example usage ===
beam = BeamSimulation(ds, n=N, L=L, dt=dt)
beam.nodes[len(beam.nodes)//2].u = 0.1  # Initial perturbation

u_hist = beam.run(400)

plt.imshow(u_hist.T, aspect='auto', origin='lower', cmap='plasma')
plt.colorbar(label='u displacement')
plt.title('Evolution of u(s,t)')
plt.xlabel('Time step')
plt.ylabel('Node index')
plt.show()

# === Example usage ===
beam = BeamSimulation(ds, n=N, L=L, dt=dt, BC_left='clamped', BC_right='clamped')

# Initial condition (midpoint displacement)
beam.nodes[len(beam.nodes)//2].u = 0.05

u_hist = beam.run(5000)

plt.imshow(u_hist.T, aspect='auto', origin='lower', cmap='plasma')
plt.colorbar(label='u displacement')
plt.title('Clamped-Free Beam: u(s,t)')
plt.xlabel('Time step')
plt.ylabel('Node index')
plt.show()
```
